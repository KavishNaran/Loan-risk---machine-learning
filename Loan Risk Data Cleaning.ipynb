# Data Cleaning

```python
#Import dependenices
import csv
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
# Specify the path to your CSV file
csv_file_path = 'Resources/credit_train.csv'

# Read the CSV file into a DataFrame
loan_data = pd.read_csv(csv_file_path)

#Print size of data set
print(loan_data.shape)

# Display the DataFrame
loan_data.head(5)
```

```python
loan_data_clean = loan_data.applymap(lambda x: x.lower() if isinstance(x, str) else x)
loan_data_clean.head(5)
```

```python
#Drop columns which are not required for analysis
loan_data_clean = loan_data_clean.drop(columns=['Loan ID','Customer ID','Years in current job', 'Loan Status', 'Months since last delinquent', 'Tax Liens', 'Purpose'])
loan_data_clean.head(5)
```

```python
#Drop all null values from data frame
loan_data_clean.dropna(inplace=True)

#Print size of data set
print(loan_data_clean.shape)
```

```python
# Verify the categories of the 'Term','Home Onwership' and 'Purpose'column
print('--------')
print(loan_data_clean['Term'].value_counts())
print('--------')
print(loan_data_clean['Home Ownership'].value_counts())
```

```python
loan_data_clean = loan_data_clean.replace({'havemortgage': 'home mortgage'})
print(loan_data_clean['Home Ownership'].value_counts())
```

```python
# Verify the categories of the 'Term' column
loan_data_clean['Term'].value_counts()
```

```python
# Transform the Home Ownership column using get_dummies
home_dummies = pd.get_dummies(loan_data_clean['Home Ownership'])

# Display the transformed data
home_dummies.tail()
```

```python
# Transform the Term column using get_dummies
term_dummies = pd.get_dummies(loan_data_clean['Term'])

# Display the transformed data
term_dummies.tail()
```

```python
# Concatenate the df_shopping_transformed and the card_dummies DataFrames
loan_data_clean = pd.concat([loan_data_clean, term_dummies], axis = 1)
loan_data_clean = pd.concat([loan_data_clean, home_dummies], axis = 1)


# Drop the original education column
loan_data_clean = loan_data_clean.drop(columns=['Term'])
loan_data_clean = loan_data_clean.drop(columns=['Home Ownership'])

# Display the DataFrame
loan_data_clean.head()
```

```python
# Capitalise the headings
loan_data_clean = loan_data_clean.rename(columns=lambda x: x.capitalize())
loan_data_clean.head(5)
```

```python
# Change Bankruptcies column into a boolean value
# Define the condition
condition = loan_data_clean['Bankruptcies'] > 1

# Create a new boolean column based on the condition
loan_data_clean['Bankrupt'] = condition

# Convert to numerical values
loan_data_clean['Bankrupt'] = loan_data_clean['Bankrupt'].astype(int)

#Drop Bankruptcies column
loan_data_clean = loan_data_clean.drop('Bankruptcies', axis = 1)

# Display the DataFrame
loan_data_clean.head()
```

```python
loan_data_clean.to_csv('Resources/Clean_train.csv', index=False)
```
